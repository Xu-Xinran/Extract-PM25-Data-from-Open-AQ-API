Interactinv with an API
https://api.openaq.org/v1/measurements?city=Paris&parameter=pm25

When queried the Open AQ Platform API, it provides air quality data for the given location. The code below pulled air quality data for Paris, looking only for the PM25 value.

This script used requests library to fetch the data, and stored the pulled data in key-value JSON format, which is cleaner and improves readability.

```python
#Step 1 start by importing the code
import requests
```

```python
#step 2 send a get request, get a response object
result=requests.get('https://api.openaq.org/v1/measurements?city=Paris&parameter=pm25')
```

```python
#step 3: get the status code..returns an integer on success
result.status_code
```

```python
#step 4: get the data
result.text
```

```python
#step 5: get data in a more readable json format
result_json=result.json()
```

```python
result_json
```

```python
# get the most recent value by assessing the first element of the list with index 0
result_json['results'][0]
```

The code above gets me the PM25 concentration in the air in Paris for March 15,2021.
